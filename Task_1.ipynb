### Report: Some of the feature are negatively correlated with the class and some has no correlation. Such features do not contribute meaningful information to the predictive power of the model and can introduce noise, leading to overfitting and decreased generalizability. By dropping these irrelevant features, we can reduce the complexity of the model, improve its training efficiency.

```python
# Categorical attributes exploration
df['EJ'].describe()
```

```python
df["EJ"].value_counts()
```

```python
dummies = pd.get_dummies(df["EJ"], prefix="EJ")
df = pd.concat([df, dummies], axis=1)
```

```python
df["EJ_A"] = df["EJ_A"].astype('float64')
df["EJ_B"] = df["EJ_B"].astype('float64')

df_numeric = df.select_dtypes(include=[np.number])
corr_matrix = df_numeric.corr()
corr_matrix["Class"].sort_values(ascending=False)
```

```python
df["EJ"] = df["EJ"].map({
    "A":0,
    "B":1
})
df["EJ"] = df["EJ"].astype('float64')
```

```python
df_numeric = df.select_dtypes(include=[np.number])
corr_matrix = df_numeric.corr()
corr_matrix["Class"].sort_values(ascending=False)
```

### Report: Since the EJ column provides a straightforward interpretation of its impact on the target variable. Keeping EJ simplifies the dataset by retaining essential information while avoiding duplication found in EJ_A and EJ_B, which serve the same purpose but split into two variables unnecessarily.

```python
df = df.drop(columns=["Id", "EJ_A", "EJ_B"]) # since id column is irrelevant in this case
```

```python
df.isna().sum()
```

```python
# two column has higher missing, imputing these column can lead to poor results
df = df.drop(columns=['BQ','EL'])
lower_missing_values_cols = [ 'CB', 'CC', 'DU', 'FC', 'FL', 'FS', 'GL']
df[lower_missing_values_cols] = df[lower_missing_values_cols].fillna(df[lower_missing_values_cols].median())
```

```python
# keeping only the higher correlated features

corr_threshold = 0.1
target_corr_abs = df.corr()['Class'].abs()

selected_features = target_corr_abs[target_corr_abs >= corr_threshold].index
df_selected = df[selected_features]
print("Selected features based on correlation threshold:", selected_features)
print("Shape of DataFrame after feature selection:", df_selected.shape)
```

```python
df_selected.describe()
```

```python
df_selected.head()
```

```python
df_selected.info()
```

```python
X = df_selected.drop(columns=["Class"], axis=1)
y = df_selected["Class"].copy()
```

### 2 Apply t-SNE algorithms

```python
from sklearn.manifold import TSNE
from sklearn.preprocessing import StandardScaler
import seaborn as sn

standardized_df = StandardScaler().fit_transform(X)

model = TSNE(n_components = 2, random_state = 42)
 
tsne_data = model.fit_transform(X)
tsne_data = np.vstack((tsne_data.T, y)).T
tsne_df = pd.DataFrame(data = tsne_data, columns =("Dim_1", "Dim_2", "label"))
 
sn.scatterplot(data=tsne_df, x='Dim_1', y='Dim_2', hue='label', palette="bright")
plt.show()
```

### 3. Apply ML algorithms

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.15, stratify=y, random_state=42)
print(X_train.shape)
print(X_valid.shape)
```

```python
# cross-validation function
from sklearn.model_selection import cross_val_score

def my_cross_validation(clf):
    print(cross_val_score(clf, X_train, y_train, cv=5, scoring='accuracy').mean())
```

```python
# Random forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import make_pipeline

rnd_clf = make_pipeline(
    StandardScaler(),
    RandomForestClassifier(n_estimators=500, max_leaf_nodes=20, n_jobs=-1, random_state=42)
)
my_cross_validation(rnd_clf)
```

```python
# SGD classifier
from sklearn.linear_model import SGDClassifier

sgd_clf = make_pipeline(
    StandardScaler(),
    SGDClassifier(loss="log_loss", alpha=0.01, max_iter=500, n_jobs=-1, random_state=42, learning_rate='adaptive', eta0=0.1, 
                  early_stopping=True, n_iter_no_change=10, warm_start=True)
)

sgd_clf.fit(X_train, y_train)
my_cross_validation(sgd_clf)
```

```python
# StackingClassifier
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

stacking_clf = make_pipeline(
    StandardScaler(),
    StackingClassifier(
        estimators=[
            ('sgd', SGDClassifier(loss="log_loss",  max_iter=500, n_jobs=-1, random_state=42, learning_rate='adaptive', eta0=0.1, 
                  early_stopping=True, n_iter_no_change=10)),
            ('lr', LogisticRegression(C=0.1, random_state=42, n_jobs=-1)),
            ('dt', DecisionTreeClassifier(max_depth=10, min_samples_leaf=5, max_features="log2", random_state=42))
        ],
        final_estimator= RandomForestClassifier(n_estimators=500, max_leaf_nodes=20, n_jobs=-1, random_state=42),
        cv= 3
    )
)
my_cross_validation(stacking_clf)
```

```python
# Precision vs Recall Curve
from sklearn.metrics import precision_recall_curve
from sklearn.model_selection import cross_val_predict

def my_precision_vs_recall_curves(clfs, clf_labels):
    plt.figure(figsize=(10, 5))

    for clf, label in zip(clfs, clf_labels):
        y_score = cross_val_predict(clf, X_train, y_train, cv=3, method="predict_proba")[:, 1]
        precision, recall, _ = precision_recall_curve(y_train, y_score)
        plt.plot(precision, recall, label=label, linewidth=2, linestyle='solid')

    plt.title("Precision vs Recall Curves")
    plt.xlabel("Recall")
    plt.ylabel("Precision")
    plt.legend()
    plt.show()
```

```python
clfs = [sgd_clf,rnd_clf, stacking_clf]
clf_labels = ["SGDClassifier","Random Forest Classifier", "Stacking Classifier"]
my_precision_vs_recall_curves(clfs, clf_labels)
```

```python
# Random forest performed better than other classifier, let'see the classification report
from sklearn.metrics import classification_report

def my_classification_report_on_train_data(clf):
    y_train_pred = cross_val_predict(clf, X_train, y_train, cv=5)
    cr = classification_report(y_train, y_train_pred)
    print(cr)

my_classification_report_on_train_data(rnd_clf)
```

#### There is room for improvement in correctly identifying instances of class 1 (lower recall), suggesting potential adjustments in the model to better capture true positives.

```python
## Fine-tuning the model with randomized search
from sklearn.pipeline import Pipeline
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint, uniform

# preprocessing steps
preprocessing = make_column_transformer(
    (StandardScaler(), make_column_selector(dtype_include=np.number))
)

# Define the full pipeline with preprocessing
full_pipeline_rf = Pipeline([
    ("preprocessing", preprocessing),
    ("rf_clf", RandomForestClassifier(random_state=42))
])

# Define hyperparameter distributions
param_distributions_rf = {
    'rf_clf__n_estimators': randint(low=50, high=200),
    'rf_clf__max_depth': randint(low=5, high=20),
    'rf_clf__min_samples_split': randint(low=2, high=20),
    'rf_clf__min_samples_leaf': randint(low=1, high=10),
    'rf_clf__max_features': ['sqrt', 'log2']
}


rnd_search_rf = RandomizedSearchCV(
    full_pipeline_rf, param_distributions=param_distributions_rf, n_iter=20, cv=3, 
    scoring='accuracy', random_state=42, verbose=1, n_jobs=-1
)
rnd_search_rf.fit(X_train, y_train)
```

```python
print("Best Parameters:", rnd_search_rf.best_params_)
print("Best Score:", rnd_search_rf.best_score_)
```

```python
# let's define the final model
final_model = rnd_search_rf.best_estimator_
final_model
```

```python
# model evaluation on validation data
y_pred = final_model.predict(X_valid)
cr = classification_report(y_valid, y_pred)
print(cr)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_valid, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Class 0', 'Class 1'])
disp.plot(cmap='Blues', values_format='d')  # 'd' for integer formatting

plt.title('Confusion Matrix')
plt.show()
```
